### **2- Accessing your dataset values:**

### **3- Display the column names of the data_frame:**

### **4- Exploring data types of each column**

Use **data_frame.info()** function to get a summary of data frame, including data type and count of not null values.

```python
# List of data types
data_frame.info()
```

### **5- Get summary of columns**

In a dataset, their can be value errors and it is important to identify them. For that we can check descriptive stats of the dataset. The **describe()** function outputs descriptive statistics.

* By default, for numeric data, the result will include **count, mean, std, min, and max, as well as lower, 50, and upper percentiles.**

* For object data (e.g. categorical values), the result will include **count, unique, top, and freq.**  

```python
#Stats for numeric data
data_frame.describe()
```

```python
# Object Stats
data_frame.describe(include=object)
```

### **6- Dataset dimension:**

Dataset dimension is number of rows and number of columns in the dataset. Normally rows are known as instances/data points and columns are known as features or variables. To find that we use **shape**.


```python
# Num of Rows and Columns in Dataset
data_frame.shape
```

### **7- Find Missing data values:**

There can be plenty of missiong values in a dataset. We get the missing (null) value count by using **isnull().sum()** functions.

```python
# Missing Values Count
data_frame.isnull().sum()
```

To find the percentage of missing data values per variable, :**Divide it by length of data set and multiply by 100%**


```python
# Missing Values Percentage
data_frame.isnull().sum()/len(data_frame)*100
```
